```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification,pipeline
modelname='google-bert/bert-base-uncased'
tokenizer=AutoTokenizer.from_pretrained(modelname)
#printing the original text
text='i love virat very much'
print('text :',text)
#converting the text into tokens
tokens=tokenizer.tokenize(text)
print('tokens :',tokens)
#converting the tokens into ids
ids=tokenizer.convert_tokens_to_ids(tokens)
print('ids :',ids)
#encodings for the text how transformer will do
encodings=tokenizer(text)
print('encodings :',encodings)
#decoder for the ids
decoded_text=tokenizer.decode(ids)
print('decoded_text :',decoded_text)
```

```python
from datasets import load_dataset
dataset=load_dataset('imdb')
dataset
```

```python
dataset["train"][0]
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSequenceClassification
tokenizer = AutoTokenizer.from_pretrained("distilbert/distilbert-base-uncased-finetuned-sst-2-english")
model = AutoModelForSequenceClassification.from_pretrained("distilbert/distilbert-base-uncased-finetuned-sst-2-english",num_labels=2)
```

```python
def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

tokenized_datasets = dataset.map(tokenize_function, batched=True),
```

```python
tokenized_datasets
```

```python
tokenized_datasets[0]['train'][0]
```

```python
# Unpack the DatasetDict from the tuple
tokenized_datasets = tokenized_datasets[0]  # This extracts the DatasetDict from the tuple
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir='./results',          # Output directory
    eval_strategy ="epoch",     # Evaluate every epoch
    learning_rate=2e-5,              # Learning rate
    per_device_train_batch_size=16,  # Batch size for training
    per_device_eval_batch_size=16,   # Batch size for evaluation
    num_train_epochs=1,              # Number of training epochs
    weight_decay=0.01,               # Strength of weight decay
)
training_args
```

```python
!pip install -U transformers accelerate
```

```python
from transformers import Trainer
trainer=Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["test"],
    tokenizer=tokenizer)
```

```python
trainer.train()
```

```python
results=trainer.evaluate()
print(results)
```
